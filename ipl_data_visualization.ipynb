TASK-1: Data Visualization on IPL Dataset

```python
# Import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset

matches = pd.read_csv("IPL_Matches_2008_2022.csv")

print("Dataset Shape:", matches.shape)

matches.head()
```

```python
# Dataset Information

matches.info()
```

```python
# Total Number of Matches Played

total_matches = matches.shape[0]

print("Total Matches Played in IPL:", total_matches)
```

```python
# Total Number of IPL Seasons

total_seasons = matches["Season"].nunique()

print("Total IPL Seasons:", total_seasons)
```

```python
# Total Matches Played in Each Season

matches_per_season = matches.groupby("Season").size()

print(matches_per_season)

plt.figure(figsize=(10, 5))
sns.barplot(x=matches_per_season.index, y=matches_per_season.values)

plt.title("Matches Played per IPL Season")
plt.xlabel("Season")
plt.ylabel("Number of Matches")

plt.show()
```

```python
# Maximum and Minimum Matches in a Season

print("Maximum Matches in a Season:", matches_per_season.max())
print("Minimum Matches in a Season:", matches_per_season.min())
```

```python
# Total Matches Won by Each Team

team_wins = matches["WinningTeam"].value_counts()

print(team_wins)

plt.figure(figsize=(12, 6))
sns.barplot(x=team_wins.index, y=team_wins.values)

plt.xticks(rotation=90)

plt.title("Total Matches Won by Each Team")
plt.xlabel("Teams")
plt.ylabel("Matches Won")

plt.show()
```

```python
# Total Matches Played in Each Stadium

stadium_matches = matches["Venue"].value_counts()

print(stadium_matches)

plt.figure(figsize=(12, 6))
sns.barplot(x=stadium_matches.index, y=stadium_matches.values)

plt.xticks(rotation=90)

plt.title("Matches Played in Each Stadium")
plt.xlabel("Stadium")
plt.ylabel("Number of Matches")

plt.show()
```

```python
# Has Toss Winning Helping in Match Winning?

toss_match = matches[matches["TossWinner"] == matches["WinningTeam"]]

percentage = (len(toss_match) / len(matches)) *100

print("Toass Winning helps in Match Winning:", percentage, "%")
```

```python
# Visualization

labels = ["Toass Win & Match Win", "Others"]

values = [
    len(toss_match),
    len(matches) - len(toss_match)
]

plt.pie(values, labels=labels, autopct='%1.1f%%')

plt.title("Impact of Toss on Match Result")

plt.show()
```

```python
#Top 15 Player of the Match Winners

top_players = matches["Player_of_Match"].value_counts().head(15)

print(top_players)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_players, y=top_players.index)

plt.title("Top 15 Player of the Match Winners")
plt.xlabel("Number of Awards")
plt.ylabel("Players")

plt.show()
```
